```
from google.colab import auth
auth.authenticate_user()

import gspread
from oauth2client.client import GoogleCredentials

gc = gspread.authorize(GoogleCredentials.get_application_default())

worksheet = gc.open('PRO 2019 Master').get_worksheet(3)

# get_all_values gives a list of rows.
rows = worksheet.get_all_values()
print(rows)

# Convert to a DataFrame and render.
import pandas as pd
PreOccupyRAW = pd.DataFrame.from_records(rows)
PreOccupyRAW = PreOccupyRAW.loc[:,5:8]
PreOccupyRAW.columns = PreOccupyRAW.loc[0,:]
PreOccupyRAW = PreOccupyRAW.drop(PreOccupyRAW.index[0])
PreOccupyRAW
```

# **整理結果**

```
PreOccupy = PreOccupyRAW
PreOccupy
```

```
#自動化報名程式

for index,row in PreOccupy.iterrows():

  data = {
      'method': 'add',
      'secDateStr': '2019-08-29',           # 考試日期
      'secId': '2',                # 上下午
      'divId': '8',                   # 組別
      'dmvNo': '43',                # 監理站代碼
      'licenseTypeCode': 'B',
      'idNo': row[0],
      'birthdayStr': row[1],
      'name': row[2],
      'contactTel': row[3]
  }
  res2 = requests.post('https://www.mvdis.gov.tw/m3-emv-trn/exm/signUp', data=data)
  res2.encoding='utf-8'
```

```
for index,row in PreOccupy.iterrows():

  data = {
      'method': 'query',
      'reservationPK': '0',
      'idNo': row[0],
      'birthdayStr': row[1]
  }
  res2 = requests.post('https://www.mvdis.gov.tw/m3-emv-trn/exm/query', data=data)
  res2.encoding='utf-8'
  table = pd.read_html(res2.text)

  try:
    df = table[8]
    print(df.loc[0,"考試日期 (星期) Date of Test"]+df.loc[0,"考試地點 Place of Test"]+',')
  

  except IndexError:
    print("No Record,")
```
