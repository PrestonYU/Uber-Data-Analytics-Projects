## 樹林監理站

```
# 樹林監理站
import datetime
now = datetime.datetime.now()

data = {
    'method': 'query',
    'licenseTypeCode': 'B',
    'expectExamDateStr': str(now.year-1911)+ddday,
    '_onlyWeekend': 'on',
    'dmvNoLv1': '40',
    'dmvNo': '40'
}
res = requests.post('https://www.mvdis.gov.tw/m3-emv-trn/exm/locations', data=data)
res.encoding='utf-8'
```

```
import pandas as pd
from lxml import etree
from bs4 import BeautifulSoup
table = pd.read_html(res.text)
df = table[5]
df = df.loc[7:,'考試日期 (星期) Date of Test':'可報名人數 Number']
df
```
