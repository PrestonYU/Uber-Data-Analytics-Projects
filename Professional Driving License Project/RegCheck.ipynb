```
from google.colab import auth
auth.authenticate_user()

import gspread
from oauth2client.client import GoogleCredentials

gc = gspread.authorize(GoogleCredentials.get_application_default())

worksheet = gc.open('PRO 2019 Master').get_worksheet(2)

# get_all_values gives a list of rows.
rows = worksheet.get_all_values()
print(rows)

# Convert to a DataFrame and render.
import pandas as pd
RegRAW = pd.DataFrame.from_records(rows)
RegRAW = RegRAW.loc[8:,4:8]
#RegRAW = RegRAW.drop(RegRAW.index[0])
RegRAW = RegRAW[RegRAW[8] != 'Y']
RegRAW = RegRAW[RegRAW[4] != '']
RegRAW = RegRAW.reset_index(drop=True)
RegRAW
```

##檢查駕駛是否符合報名資格

```
# 假報名檢查法

#RegA = RegRAW[RegRAW['Status'] == 'Available']

import datetime
now = datetime.datetime.now()

for index,row in RegRAW.iterrows():

  data = {
      'method': 'add',
      'secDateStr': '2019-09-05',           # 考試日期
      'secId': '2',                         # 上下午
      'divId': '9',                         # 組別
      'dmvNo': '25',                        # 監理站代碼
      'licenseTypeCode': 'B',
      'idNo': row[4],
      'birthdayStr': row[5],
      'name': row[6],
      'contactTel': row[7]
  }
  res2 = requests.post('https://www.mvdis.gov.tw/m3-emv-trn/exm/signUp', data=data)
  res2.encoding='utf-8'
```

```
for index,row in RegRAW.iterrows():

  data = {
      'method': 'query',
      'reservationPK': '0',
      'idNo': row[4],
      'birthdayStr': row[5]
  }
  res2 = requests.post('https://www.mvdis.gov.tw/m3-emv-trn/exm/query', data=data)
  res2.encoding='utf-8'
  table = pd.read_html(res2.text)

  try:
    df = table[8]
    print(df.loc[0,"考試日期 (星期) Date of Test"]+df.loc[0,"考試地點 Place of Test"]+',')
    

  except IndexError:
    print('N')
```
